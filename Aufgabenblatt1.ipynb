### 5) Erläutern Sie das Ergebnis - was bedeuten die einzelnen Zeilen?

$\textit{count}$: Anzahl der Zeilen im Datensatz in denen die Spalte vorkommt

$\textit{mean}$: Der Durchschnittswert pro Spalte im Datensatz

$\textit{std}$: Standardabweichung für die jeweilige Spalte im Datensatz

$\textit{min}$: Der kleinste Wert pro Spalte im Datensatz

$\textit{25%}$: 25%-Quantile pro Spalte im Datensatz, d.h. 25% aller Werte sind kleiner als dieser Wert(von $numpy.percentile$)

$\textit{50%}$: 50%-Quantile pro Spalte im Datensatz, d.h. 50% aller Werte sind kleiner als dieser Wert (entspricht dem Median-Wert)

$\textit{75%}$: 75%-Quantile pro Spalte im Datensatz, d.h. 75% aller Werte sind kleiner als dieser Wert

$\textit{max}$: Der größte Wert pro Spalte im Datensatz

### 8) Was macht $scatter\_matrix()$?

Trägt alle Spalten gegeneinander in einem Scatter-Plot auf (aus den ersten 10 Spalten in `df_part` haben 8 numerische Werte, deswegen 8x8 = 64 Plots).

Auf der Diagonalen, wo normalerweise jede Spalte gegen sich selbst aufgetragen werden würde, wird stattdessen ein Dichte-Plot, d.h. die Verteilung der Werte, als Histogramm dargestellt. Die Höhe des Histogramms gibt die Anzahl der Werte (y-Achse) mit diesem Wert (x-Achse) an.
