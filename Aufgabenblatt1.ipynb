### 5) Erläutern Sie das Ergebnis - was bedeuten die einzelnen Zeilen?

$\textit{count}$: Anzahl der Zeilen im Datensatz in denen die Spalte vorkommt (Anzahl der Datenpunkte)

$\textit{mean}$: Der Durchschnittswert pro Spalte im Datensatz (Summe über alle Zeile / Anzahl der Zeilen)

$\textit{std}$: Standardabweichung für die jeweilige Spalte im Datensatz (Wurzel der Varianz)

$\textit{min}$: Der kleinste Wert pro Spalte im Datensatz

$\textit{25%}$: 25%-Quantile pro Spalte im Datensatz, d.h. 25% aller Werte sind kleiner gleich als dieser Wert

$\textit{50%}$: 50%-Quantile pro Spalte im Datensatz, d.h. 50% aller Werte sind kleiner gleich als dieser Wert (entspricht dem Median-Wert)

$\textit{75%}$: 75%-Quantile pro Spalte im Datensatz, d.h. 75% aller Werte sind kleiner gleich als dieser Wert

$\textit{max}$: Der größte Wert pro Spalte im Datensatz

### 8) Was macht $scatter\_matrix()$?

Trägt alle Spalten gegeneinander in einem Scatter-Plot auf (aus den ersten 10 Spalten in `df_part` haben 8 numerische Werte, deswegen 8x8 = 64 Plots).

Auf der Diagonalen, wo normalerweise jede Spalte gegen sich selbst aufgetragen werden würde, wird stattdessen die Verteilung der Werte als Histogramm dargestellt. Die Höhe des Histogramms gibt die Anzahl der Werte (y-Achse) mit diesem Wert (x-Achse) an.

### 9) Was bedeuten die Parameter $figsize=(15, 15)$, $s=50$ und $marker=D$?

$figsize=(15,15)$: Die Größe des Plots in Inches

$s=50$: Größe der einzelnen Scatter-Punkte

$marker=D$: Art der Darstellung der Daten im Plot (Dreieck, Kreis, Quadrat, Raute)

```python
pandas.plotting.scatter_matrix(df_part, figsize=(15,15), s=5, marker='D')
plt.show()
```

### 10) Erstellen Sie mit $matplotlib.boxplot$ einen Boxplot des Features $Outstate$

```python
plt.boxplot(df["Outstate"])
plt.show()
```

### 11) Interpretieren Sie den Boxplot

Zwischen dem untersten (minimum = 25%-Quantile - $1.5\times$IQR) und dem obersten waagerechten Strich (maximum = 75%-Quantile + 1.5$\times$IQR) liegen 99.3% aller Daten aus dem Datensatz. Die sog. `Outliers`, der Kreis am oberen Ende des Plots, markieren die letzten 0.7% der Daten.

Das untere Ende des Rechtecks markiert die 25%-Quantile, das obere die 75%-Quantile (`interquartile Range, IQR`). Der rote, waagerechte Strich im Rechteck markiert die 50%-Quantile (`Median`).

Der Abstandsfaktor `1.5` kann mit dem Paramter `wisk` umgestellt werden. Sollte der Datenpunkt bei (25% - 1.5*IQR) bzw. (75% + 1.5*IQR) nicht existieren wird der Balken auf den nächst-nähere Punkt gesetzt.
